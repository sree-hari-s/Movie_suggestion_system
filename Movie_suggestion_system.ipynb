```python
import ast
def convert(obj):
    ls = []
    for data in ast.literal_eval(obj):
        ls.append(data['name'])
    return ls
```

```python
def convert_cast(obj):
    ls = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            ls.append(i['character'])
            count +=1
        else:
            break
    return ls
            
        
        
```

```python
 movies['cast'].apply(convert_cast)
```

```python
 movies['cast'] =  movies['cast'].apply(convert_cast)
```

```python
def find_director(obj):
    L = []
    for data in ast.literal_eval(obj):
        if data['job'] == "Director":
            L.append(data['name'])
    return L
        
print(find_director(movies.crew.iloc[0]))
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

### Text Vectorization

Text Vectorization is the process of converting text into numerical representation. Here is some popular methods to accomplish text vectorization:

- Binary Term Frequency
- Bag of Words (BoW) Term Frequency
- (L1) Normalized Term Frequency
- (L2) Normalized TF-IDF
- Word2Vec

Read futher about [text vectorization](https://towardsdatascience.com/getting-started-with-text-vectorization-2f2efbec6685#:~:text=Text%20Vectorization%20is%20the%20process,(L1)%20Normalized%20Term%20Frequency)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tag']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df['tag'][0]
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
print(stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d jakesully neytiri dr.graceaugustine jamescameron'))
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
#In high dimension we use 'consine distance' rather than Euclidean distance.
## Eculedian distance fail on high dimensions 
## so we use cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
# we get a similarity matrix of 4806 items
#which give similarity when an idex positon is given
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key= lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True, key= lambda x: x[1])[1:6]
    for data in movies_list:
        print(new_df.iloc[data[0]].title)
    

recommend("John Carter")
```
